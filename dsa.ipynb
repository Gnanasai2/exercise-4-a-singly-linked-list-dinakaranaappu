```python
class Stack:
    def __init__(self):
        # Initialize an empty list to represent the stack
        self.stack = []

    def is_empty(self):
        # Check if the stack is empty
        return len(self.stack) == 0

    def push(self, item):
        # Add an item to the top of the stack
        self.stack.append(item)

    def pop(self):
        # Remove the top item from the stack and return it
        if not self.is_empty():
            return self.stack.pop()
        else:
            return "Stack is empty!"

    def peek(self):
        # Get the top item of the stack without removing it
        if not self.is_empty():
            return self.stack[-1]
        else:
            return "Stack is empty!"

    def size(self):
        # Return the number of items in the stack
        return len(self.stack)

    def display(self):
        # Display the elements of the stack
        if self.is_empty():
            return "Stack is empty!"
        else:
            return self.stack


def menu():
    print("\nStack Operations Menu:")
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Size")
    print("5. Display")
    print("6. Exit")

def main():
    stack = Stack()

    while True:
        menu()  # Display the menu
        try:
            choice = int(input("Choose an operation (1-6): "))
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 6.")
            continue

        if choice == 1:
            item = input("Enter the value to push: ")
            stack.push(item)
            print(f"'{item}' has been pushed onto the stack.")

        elif choice == 2:
            popped_item = stack.pop()
            print(f"Popped item: {popped_item}")

        elif choice == 3:
            top_item = stack.peek()
            print(f"Top element: {top_item}")

        elif choice == 4:
            print(f"Size of stack: {stack.size()}")

        elif choice == 5:
            print("Stack elements:", stack.display())

        elif choice == 6:
            print("Exiting program.")
            break

        else:
            print("Invalid choice! Please select a valid operation.")

if __name__ == "__main__":
    main()
```

```python
class LinearQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size  # initialize an empty list of fixed size
        self.front = -1  # to point to the front of the queue
        self.rear = -1   # to point to the rear of the queue

    def is_empty(self):
        # Check if the queue is empty
        return self.front == -1

    def is_full(self):
        # Check if the queue is full
        return self.rear == self.size - 1

    def enqueue(self, item):
        # Add an item to the queue
        if self.is_full():
            print("Queue is full!")
        else:
            if self.front == -1:  # If the queue is empty, initialize front
                self.front = 0
            self.rear += 1
            self.queue[self.rear] = item
            print(f"Enqueued: {item}")

    def dequeue(self):
        # Remove and return the front item from the queue
        if self.is_empty():
            print("Queue is empty!")
        else:
            item = self.queue[self.front]
            if self.front == self.rear:  # Only one element in the queue
                self.front = self.rear = -1  # Reset queue to empty
            else:
                self.front += 1
            return item

    def get_front(self):
        # Get the front element without removing it
        if self.is_empty():
            return "Queue is empty!"
        else:
            return self.queue[self.front]

    def display(self):
        # Display the elements of the queue
        if self.is_empty():
            print("Queue is empty!")
        else:
            print("Queue elements: ", self.queue[self.front:self.rear+1])


# Example usage of LinearQueue
linear_queue = LinearQueue(5)
linear_queue.enqueue(10)
linear_queue.enqueue(20)
linear_queue.enqueue(30)
linear_queue.display()  # Output: [10, 20, 30]
print("Dequeued:", linear_queue.dequeue())  # Output: 10
linear_queue.display()  # Output: [20, 30]
print("Front element:", linear_queue.get_front())  # Output: 20
```

```python
class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size  # initialize an empty list of fixed size
        self.front = -1  # to point to the front of the queue
        self.rear = -1   # to point to the rear of the queue

    def is_empty(self):
        # Check if the queue is empty
        return self.front == -1

    def is_full(self):
        # Check if the queue is full (if rear is just behind front)
        return (self.rear + 1) % self.size == self.front

    def enqueue(self, item):
        # Add an item to the queue
        if self.is_full():
            print("Queue is full!")
        else:
            if self.front == -1:  # If the queue is empty, initialize front and rear
                self.front = 0
            self.rear = (self.rear + 1) % self.size
            self.queue[self.rear] = item
            print(f"Enqueued: {item}")

    def dequeue(self):
        # Remove and return the front item from the queue
        if self.is_empty():
            print("Queue is empty!")
        else:
            item = self.queue[self.front]
            if self.front == self.rear:  # Only one element in the queue
                self.front = self.rear = -1  # Reset the queue to empty
            else:
                self.front = (self.front + 1) % self.size
            return item

    def get_front(self):
        # Get the front element without removing it
        if self.is_empty():
            return "Queue is empty!"
        else:
            return self.queue[self.front]

    def display(self):
        # Display the elements of the queue
        if self.is_empty():
            print("Queue is empty!")
        else:
            i = self.front
            print("Queue elements: ", end="")
            while True:
                print(self.queue[i], end=" ")
                if i == self.rear:
                    break
                i = (i + 1) % self.size
            print()


# Example usage of CircularQueue
circular_queue = CircularQueue(5)
circular_queue.enqueue(10)
circular_queue.enqueue(20)
circular_queue.enqueue(30)
circular_queue.display()  # Output: 10 20 30
print("Dequeued:", circular_queue.dequeue())  # Output: 10
circular_queue.display()  # Output: 20 30
print("Front element:", circular_queue.get_front())  # Output: 20
```

```python
#single linked list
class Node:
    def __init__(self, data):
        self.data = data  # Data of the node
        self.next = None  # Reference to the next node (initially None)

class SinglyLinkedListStack:
    def __init__(self):
        self.top = None  # Initialize stack as empty (no nodes)

    def is_empty(self):
        return self.top is None  # Stack is empty if top is None

    def push(self, data):
        # Create a new node with the given data
        new_node = Node(data)
        new_node.next = self.top  # Make the new node point to the current top
        self.top = new_node  # Move the top pointer to the new node
        print(f"Pushed: {data}")

    def pop(self):
        if self.is_empty():
            print("Stack is empty!")
            return None
        else:
            popped_data = self.top.data  # Get the data from the top node
            self.top = self.top.next  # Move the top pointer to the next node
            print(f"Popped: {popped_data}")
            return popped_data

    def peek(self):
        if self.is_empty():
            print("Stack is empty!")
            return None
        else:
            print(f"Top element: {self.top.data}")
            return self.top.data

    def display(self):
        if self.is_empty():
            print("Stack is empty!")
        else:
            current = self.top
            print("Stack elements: ", end="")
            while current:
                print(current.data, end=" ")
                current = current.next
            print()  # Newline


# Example usage of SinglyLinkedListStack
singly_stack = SinglyLinkedListStack()
singly_stack.push(10)
singly_stack.push(20)
singly_stack.push(30)
singly_stack.display()  # Output: 30 20 10
singly_stack.pop()  # Output: Popped: 30
singly_stack.peek()  # Output: Top element: 20
singly_stack.display()  # Output: 20 10
```

```python
#doubly linked list

class DoublyNode:
    def __init__(self, data):
        self.data = data  # Data of the node
        self.next = None  # Reference to the next node (initially None)
        self.prev = None  # Reference to the previous node (initially None)

class DoublyLinkedListStack:
    def __init__(self):
        self.top = None  # Initialize stack as empty (no nodes)

    def is_empty(self):
        return self.top is None  # Stack is empty if top is None

    def push(self, data):
        # Create a new node with the given data
        new_node = DoublyNode(data)
        new_node.next = self.top  # Make the new node point to the current top
        if self.top is not None:
            self.top.prev = new_node  # Update the previous reference of the old top
        self.top = new_node  # Move the top pointer to the new node
        print(f"Pushed: {data}")

    def pop(self):
        if self.is_empty():
            print("Stack is empty!")
            return None
        else:
            popped_data = self.top.data  # Get the data from the top node
            self.top = self.top.next  # Move the top pointer to the next node
            if self.top is not None:
                self.top.prev = None  # Reset the previous reference of the new top
            print(f"Popped: {popped_data}")
            return popped_data

    def peek(self):
        if self.is_empty():
            print("Stack is empty!")
            return None
        else:
            print(f"Top element: {self.top.data}")
            return self.top.data

    def display(self):
        if self.is_empty():
            print("Stack is empty!")
        else:
            current = self.top
            print("Stack elements (top to bottom): ", end="")
            while current:
                print(current.data, end=" ")
                current = current.next
            print()  # Newline


# Example usage of DoublyLinkedListStack
doubly_stack = DoublyLinkedListStack()
doubly_stack.push(10)
doubly_stack.push(20)
doubly_stack.push(30)
doubly_stack.display()  # Output: 30 20 10
doubly_stack.pop()  # Output: Popped: 30
doubly_stack.peek()  # Output: Top element: 20
doubly_stack.display()  # Output: 20 10
```

```python
#stack using linked list
class Node:
    def __init__(self, data):
        self.data = data  # Data in the node
        self.next = None  # Reference to the next node

class Stack:
    def __init__(self):
        self.top = None  # Initialize an empty stack

    def is_empty(self):
        # Stack is empty if top is None
        return self.top is None

    def push(self, data):
        # Create a new node
        new_node = Node(data)
        new_node.next = self.top  # Make new node point to the current top
        self.top = new_node  # Move top to the new node
        print(f"Pushed: {data}")

    def pop(self):
        # Remove the top item from the stack
        if self.is_empty():
            print("Stack is empty!")
            return None
        else:
            popped_data = self.top.data  # Get the data from the top node
            self.top = self.top.next  # Move top to the next node
            print(f"Popped: {popped_data}")
            return popped_data

    def peek(self):
        # Return the top element without removing it
        if self.is_empty():
            print("Stack is empty!")
            return None
        else:
            return self.top.data

    def display(self):
        # Display all elements in the stack
        if self.is_empty():
            print("Stack is empty!")
        else:
            current = self.top
            print("Stack elements (top to bottom): ", end="")
            while current:
                print(current.data, end=" ")
                current = current.next
            print()  # Newline


# Example usage of Stack
stack = Stack()
stack.push(10)
stack.push(20)
stack.push(30)
stack.display()  # Output: 30 20 10
stack.pop()  # Output: Popped: 30
stack.peek()  # Output: 20
stack.display()  # Output: 20 10
```

```python
#queue using linked list

class Node:
    def __init__(self, data):
        self.data = data  # Data in the node
        self.next = None  # Reference to the next node

class Queue:
    def __init__(self):
        self.front = None  # Front of the queue
        self.rear = None   # Rear of the queue

    def is_empty(self):
        # Queue is empty if front is None
        return self.front is None

    def enqueue(self, data):
        # Create a new node
        new_node = Node(data)
        if self.is_empty():
            self.front = self.rear = new_node  # If queue is empty, both front and rear point to the new node
        else:
            self.rear.next = new_node  # Add new node at the rear of the queue
            self.rear = new_node  # Move rear to the new node
        print(f"Enqueued: {data}")

    def dequeue(self):
        # Remove the front item from the queue
        if self.is_empty():
            print("Queue is empty!")
            return None
        else:
            dequeued_data = self.front.data  # Get the data from the front node
            self.front = self.front.next  # Move front to the next node
            if self.front is None:  # If queue becomes empty after dequeue, reset rear
                self.rear = None
            print(f"Dequeued: {dequeued_data}")
            return dequeued_data

    def peek(self):
        # Return the front element without removing it
        if self.is_empty():
            print("Queue is empty!")
            return None
        else:
            return self.front.data

    def display(self):
        # Display all elements in the queue
        if self.is_empty():
            print("Queue is empty!")
        else:
            current = self.front
            print("Queue elements (front to rear): ", end="")
            while current:
                print(current.data, end=" ")
                current = current.next
            print()  # Newline


# Example usage of Queue
queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
queue.display()  # Output: 10 20 30
queue.dequeue()  # Output: Dequeued: 10
queue.peek()  # Output: 20
queue.display()  # Output: 20 30
```

```python
#selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        # Find the index of the minimum element in the unsorted portion of the array
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Swap the found minimum element with the first unsorted element
        arr[i], arr[min_index] = arr[min_index], arr[i]

        print(f"Step {i + 1}: {arr}")  # Display the array after each step

# Example usage of Selection Sort
arr = [64, 25, 12, 22, 11]
print("Original Array:", arr)
selection_sort(arr)
print("Sorted Array:", arr)
```

```python
#insertion sort
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]  # The element to be inserted
        j = i - 1

        # Move elements of arr[0..i-1] that are greater than key
        # to one position ahead of their current position
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        # Insert the key at the correct position
        arr[j + 1] = key

        print(f"Step {i}: {arr}")  # Display the array after each step

# Example usage of Insertion Sort
arr = [64, 25, 12, 22, 11]
print("Original Array:", arr)
insertion_sort(arr)
print("Sorted Array:", arr)
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key    # The key of the node
        self.value = value  # The value associated with the key
        self.next = None  # Next node in the chain (linked list)

class HashTableChaining:
    def __init__(self, size):
        self.size = size  # Size of the hash table
        self.table = [None] * self.size  # Initialize the hash table with None

    def hash_function(self, key):
        # Simple hash function to get index in the table
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        new_node = Node(key, value)

        # Handle collision with chaining (linked list)
        if self.table[index] is None:
            self.table[index] = new_node  # If no collision, just insert the node
        else:
            current = self.table[index]
            while current.next:
                current = current.next
            current.next = new_node  # Append new node at the end of the chain
        print(f"Inserted: ({key}, {value})")

    def search(self, key):
        index = self.hash_function(key)
        current = self.table[index]

        # Traverse the linked list at that index to find the key
        while current:
            if current.key == key:
                return current.value  # Key found, return its value
            current = current.next
        return None  # Key not found

    def delete(self, key):
        index = self.hash_function(key)
        current = self.table[index]
        prev = None

        while current:
            if current.key == key:
                if prev:
                    prev.next = current.next  # Bypass the current node
                else:
                    self.table[index] = current.next  # Remove from the start of the chain
                print(f"Deleted: ({key}, {current.value})")
                return
            prev = current
            current = current.next
        print(f"Key {key} not found!")

    def display(self):
        for i, node in enumerate(self.table):
            if node:
                chain = []
                current = node
                while current:
                    chain.append(f"({current.key}, {current.value})")
                    current = current.next
                print(f"Index {i}: {', '.join(chain)}")
            else:
                print(f"Index {i}: Empty")

# Example usage of HashTable with Chaining
ht = HashTableChaining(10)
ht.insert("apple", 100)
ht.insert("banana", 200)
ht.insert("grape", 300)
ht.insert("orange", 400)
ht.display()

print("Search for 'banana':", ht.search("banana"))
ht.delete("banana")
ht.display()
```

```python
# merge sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    # Merge the sorted halves
    return merge(left_half, right_half)

def merge(left, right):
    sorted_arr = []
    i = j = 0

    # Merge two sorted arrays
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            sorted_arr.append(right[j])
            j += 1

    # If there are any remaining elements in left or right, add them
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])

    return sorted_arr

# Example usage of Merge Sort
arr = [38, 27, 43, 3, 9, 82, 10]
print("Original Array:", arr)
sorted_arr = merge_sort(arr)
print("Sorted Array:", sorted_arr)
```

```python
# quick sort

def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    # Choose the pivot (using the last element here)
    pivot = arr[-1]
    left = [x for x in arr[:-1] if x <= pivot]  # Elements less than or equal to pivot
    right = [x for x in arr[:-1] if x > pivot]  # Elements greater than pivot

    # Recursively sort the left and right sub-arrays and combine with pivot
    return quick_sort(left) + [pivot] + quick_sort(right)

# Example usage of Quick Sort
arr = [38, 27, 43, 3, 9, 82, 10]
print("Original Array:", arr)
sorted_arr = quick_sort(arr)
print("Sorted Array:", sorted_arr)
```

```python
# binary search tree

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.value = key

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # Insert a new node with given key
    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert(self.root, key)

    def _insert(self, root, key):
        # Recursively insert in the right place based on the BST property
        if key < root.value:
            if root.left is None:
                root.left = Node(key)
            else:
                self._insert(root.left, key)
        elif key > root.value:
            if root.right is None:
                root.right = Node(key)
            else:
                self._insert(root.right, key)

    # Search for a node with the given key
    def search(self, key):
        return self._search(self.root, key)

    def _search(self, root, key):
        # Base cases: root is None or key is present at the root
        if root is None or root.value == key:
            return root
        # If key is greater than the root's key, search in the right subtree
        if key > root.value:
            return self._search(root.right, key)
        # Otherwise, search in the left subtree
        return self._search(root.left, key)

    # In-order traversal (left, root, right)
    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, root, result):
        if root:
            self._inorder(root.left, result)
            result.append(root.value)
            self._inorder(root.right, result)

    # Pre-order traversal (root, left, right)
    def preorder(self):
        result = []
        self._preorder(self.root, result)
        return result

    def _preorder(self, root, result):
        if root:
            result.append(root.value)
            self._preorder(root.left, result)
            self._preorder(root.right, result)

    # Post-order traversal (left, right, root)
    def postorder(self):
        result = []
        self._postorder(self.root, result)
        return result

    def _postorder(self, root, result):
        if root:
            self._postorder(root.left, result)
            self._postorder(root.right, result)
            result.append(root.value)

    # Delete a node from the BST
    def delete(self, key):
        self.root = self._delete(self.root, key)

    def _delete(self, root, key):
        # Base case: if the tree is empty
        if root is None:
            return root

        # Find the node to be deleted
        if key < root.value:
            root.left = self._delete(root.left, key)
        elif key > root.value:
            root.right = self._delete(root.right, key)
        else:
            # Node with one child or no child
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left

            # Node with two children: Get the inorder successor (smallest in the right subtree)
            min_node = self._min_value_node(root.right)
            root.value = min_node.value  # Copy the inorder successor's content to this node
            root.right = self._delete(root.right, min_node.value)  # Delete the inorder successor

        return root

    def _min_value_node(self, node):
        # Find the node with the smallest value (leftmost node)
        current = node
        while current.left:
            current = current.left
        return current

# Example usage of the Binary Search Tree (BST)
if __name__ == "__main__":
    bst = BinarySearchTree()

    # Inserting nodes
    bst.insert(50)
    bst.insert(30)
    bst.insert(20)
    bst.insert(40)
    bst.insert(70)
    bst.insert(60)
    bst.insert(80)

    print("In-order Traversal:", bst.inorder())   # Should print [20, 30, 40, 50, 60, 70, 80]
    print("Pre-order Traversal:", bst.preorder()) # Should print [50, 30, 20, 40, 70, 60, 80]
    print("Post-order Traversal:", bst.postorder()) # Should print [20, 40, 30, 60, 80, 70, 50]

    # Searching for a value
    key = 40
    node = bst.search(key)
    if node:
        print(f"Node with key {key} found!")
    else:
        print(f"Node with key {key} not found!")

    # Deleting a node
    bst.delete(20)
    print("In-order Traversal after deleting 20:", bst.inorder()) # Should print [30, 40, 50, 60, 70, 80]

    bst.delete(30)
    print("In-order Traversal after deleting 30:", bst.inorder()) # Should print [40, 50, 60, 70, 80]
```
